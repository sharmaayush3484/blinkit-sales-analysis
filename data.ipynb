```python
import pandas as pd 
import matplotlib.pyplot as plt
df = pd.read_csv('BlinkIT-Grocery-Data.csv')
```

```python
df.info()
df.head(10)
df.describe()
```

```python
df.isnull().sum()
df = df.fillna(df.mean(numeric_only=True))
df.isnull().sum()
```

```python
df.duplicated().sum()
df = df.drop_duplicates()
df.duplicated().sum()
```

```python
df['Item Fat Content'].value_counts()
df['Item Fat Content'] = df['Item Fat Content'].replace({
    'Low Fat': 'Low Fat',
    'LF': 'Low Fat',
    'low fat': 'Low Fat',
    'Regular': 'Regular',
    'reg': 'Regular'
})
df['Item Fat Content'].value_counts()
```

```python
df.describe()
df.head(10)
```

```python
df.to_csv('BlinkIT-Grocery-Data-Cleaned.csv', index=False)
```

```python
df['Item Type'].value_counts()
```

```python
df.groupby('Item Type')['Sales'].mean().sort_values(ascending=False)
```

```python
df.groupby('Outlet Location Type')['Sales'].mean().sort_values(ascending=False)
```

```python
df.groupby('Item Fat Content')['Sales'].mean().sort_values(ascending=False)
```

```python
df.groupby('Item Type')['Sales'].sum().sort_values(ascending=False).plot(kind='barh')
plt.title('Total Sales by Item Type')
plt.xlabel('Total Sales')
plt.ylabel('Item Type')
plt.show()
```

```python
df.groupby('Outlet Location Type')['Sales'].sum().sort_values(ascending=False).plot(kind='barh')
plt.title('Total Sales by Outlet Location Type')
plt.xlabel('Total Sales')
plt.ylabel('Outlet Location Type')
plt.show()
```

```python
df.groupby('Item Fat Content')['Sales'].sum().sort_values(ascending=False).plot(kind='pie', autopct='%1.1f%%')
plt.title('Total Sales by Item Fat Content')
plt.xlabel('Total Sales')
plt.ylabel('Item Fat Content')
plt.show()
```
